```python
# Insight of pos/ neg
data_insight = []
for m in df.month.unique():
    pos = len(df[df['month'] == m][df['type'] == 'pos'])
    neg = len(df[df['month'] == m][df['type'] == 'neg'])
    dct = {}
    dct['month'] = m
    dct['pos'] = pos
    dct['neg'] = neg
    data_insight.append(dct)
df_insight = pd.DataFrame(sorted(data_insight, key= lambda x: x['month']))
```
