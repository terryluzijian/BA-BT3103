```python
# Initial, assume all important data is lebaled as '[]'
keywords = []
for t in result:
    if t['text'][0] == '[':
        tw = t['text'][t['text'].find('[')+1:t['text'].find(']')]
        if tw not in keywords:
            keywords.append(tw)
            
```

```python
keywords
```

```python
# Assign tags for all tweets using keywords, checking non-tag tweet
tags = []
non_tags = []
non_tags
ts = df.text.tolist()
for t in ts:
    is_included = False
    dct = {}
    dct['text'] = t
    dct['type'] = ['non',]
    for d in keywords:
        if d in t:
            dct['type'].append(d)
    if dct['type'] == ['non',]:
        non_tags.append(dct)
    tags.append(dct)
```

```python
# Intial of NLP training set, given Pos as taged, Neg as non_tag
# Assuming all tweet using '[]' is important content will be 

# The base assumption of nlp is: given [] is positive content, try to predict positive even with out '[]'
for t in result:
    if '[' in t['text']:
        t['type'] = 'pos'
    else:
        t['type'] = 'neg'
```

```python
# Timestamp cleanse
for t in result:
    t['month'] = t['time'].month + (t['time'].year - 2012) * 12
df = pd.DataFrame(result)
# df.to_json('Twitter.json')
```

```python
# Insight of pos/ neg
data_insight = []
for m in df.month.unique():
    pos = len(df[df['month'] == m][df['type'] == 'pos'])
    neg = len(df[df['month'] == m][df['type'] == 'neg'])
    dct = {}
    dct['month'] = m
    dct['pos'] = pos
    dct['neg'] = neg
    data_insight.append(dct)
df_insight = pd.DataFrame(sorted(data_insight, key= lambda x: x['month']))
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(20, 10))
x = df_insight.month.tolist()
y1 = df_insight.pos.tolist()
y2 = df_insight.neg.tolist()
ax.stackplot(x, y1, y2)
plt.show()
```
